```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

df = pd.read_csv('epa-sea-level.csv',header=0)
x = df['Year']
y = df['CSIRO Adjusted Sea Level']
slope, intercept, r, p, se = linregress(x, y)
plt.figure(figsize=(25,17))
plt.scatter(x , y)
xpr = pd.Series([i for i in range(1880,2050)])
ypr = xpr*slope+intercept
plt.plot(xpr,ypr,color='red')
plt.title("Rise in Sea Level", fontsize=25)
plt.xlabel("Year",fontsize=20,labelpad=13)
plt.ylabel("Sea Level (inches)",fontsize=20,labelpad=13)

df2 = df.loc[df['Year']>=2000]
x2 = df2['Year']
y2 = df2['CSIRO Adjusted Sea Level']
slope2, intercept2, r2, p2, se2 = linregress(x2, y2)
xpr2 = pd.Series([i for i in range(2000,2050)])
ypr2 = xpr2*slope2+intercept2
plt.plot(xpr2,ypr2,color='green')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
